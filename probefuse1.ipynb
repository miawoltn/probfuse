```python
hr_data_df = pd.DataFrame(data=hr_data)
hr_data_transposed = hr_data_df.T
```

```python
hr_data_transposed.columns = hr_data_transposed.iloc[0]
hr_data_transposed = hr_data_transposed[1:]
hr_data_transposed
```

```python
ir_a = hr_data_transposed[['A1', 'A2', 'A3']] #pd.concat([hr_data_transposed.A1, hr_data_transposed.A2, hr_data_transposed.A3])
```

```python
ir_a
```

```python
def segment(array, segment):
    k, m = divmod(len(array), segment)
    return (array[i * k + min(i, m):(i + 1) * k + min(i + 1, m)] for i in range(segment))
```

```python
seg_no = int(input('Enter number of segment: '))
```

```python
segments = segment(ir_a[1:], seg_no)
```
